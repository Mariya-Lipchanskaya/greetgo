```python
import json
import os
import ijson

import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline

import seaborn as sns
import tqdm

import warnings
warnings.filterwarnings('ignore')
```

```python
#Создание пустого датафрейма со столбцами collaterals 
collateral_columns = [ "no", "status", "address" ,  "collateral_id", "collateral_type", "contract_id", "description",
         "insurance_company", "mortgagor", "percentage", "summa", "summa_nt"]

collateral_data = pd.DataFrame(columns= collateral_columns)
```

```python
#Функция по считыванию ".json" файлов и добавлению в датафрейм данных о collaterals
def files(path):
    fds = sorted(os.listdir(path))
    filenames = []
    for file in fds:
        if file.endswith(('.json')): 
            filenames.append(file)
    return filenames

path = '/Users/aigulsadykova/Downloads/greetgo/greetgo_data/cs2018-07-17'

passed = 0
added = 0
filenames = files(path)


k=0
c=0

for file in filenames:
    f = open(path + '/' + file)
    data = json.load(f)
    f.close()
   
    for credit in data['credit']:
        
        c_data = credit['collateral']
        
        if c_data is None: 
            print(file, ' пропускаю')
            passed += 1
        else:
            print(file, ' добавляю')
            collateral_data = collateral_data.append(c_data)
            added += 1
```

```python
contracts = list(new_df["contract_id"].unique())
print(len(contracts))
```

```python
#Создаем новый датафрейм с размерностью 9575х2
col_data = pd.DataFrame(np.empty((9575,2)),columns = [ "contract_id","collaterals"])
```

```python
from tqdm import tqdm_notebook
```

```python
#Запускаем цикл для записи в датафрейм col_data уникального "contract_id" и соответствующее кол-во залогов
a=0
for contract in tqdm_notebook(contracts):
    temp_data = new_df[new_df['contract_id'] == contract]
    collaterals = len(list(temp_data['collateral_id'].value_counts().index))
    col_data["collaterals"].iloc[a] = collaterals
    col_data["contract_id"].iloc[a] = str(contract)
    a+=1
    
```

```python
col_data["contract_id"] = col_data["contract_id"].astype('int64')
```

```python
#Объединяем два датафрейма по 'contract_id'
collaterals = pd.merge(new_df, col_data, on='contract_id', how = "inner", indicator=False)
```

```python
collaterals.shape
```

```python
contract=collaterals.contract_id.unique()                                       
new=collaterals[["contract_id","summa_nt",'Вспомогательное оборудование', 'Гарантия','Гарантия резервных товаров', 'Денежные средства',
        'Драгоценные металлы', 'Имущество ответственности','Крупно-рогатый скот', 'Недвижимость', 'Поручитель',
        'Сберегательный депозит', 'Стоимость сельхоз техники','Уборка будущего урожая']].iloc[0:9575]
```

```python
#Записываем в одну строку данные по одному 'contract_id', суммируем по столбцам "summa_nt" и категориальным 
counter=0
for i in tqdm_notebook(contract):
    a=collaterals[collaterals.contract_id==i]
    summa=0
    b=0
    for j in range(len(a)):
        summa=int(a.summa_nt.iloc[j])+summa
        b=(a[['Вспомогательное оборудование', 'Гарантия','Гарантия резервных товаров', 'Денежные средства',
        'Драгоценные металлы', 'Имущество ответственности','Крупно-рогатый скот', 'Недвижимость', 'Поручитель',
        'Сберегательный депозит', 'Стоимость сельхоз техники','Уборка будущего урожая']].iloc[j])+b
    new["summa_nt"].iloc[counter]=summa
    new["contract_id"].iloc[counter]=i
    new["summa_nt"].iloc[counter]=summa
    new["contract_id"].iloc[counter]=i
    new['Вспомогательное оборудование'].iloc[counter]= b[0]
    new['Гарантия'].iloc[counter]= b[1]
    new['Гарантия резервных товаров'].iloc[counter]= b[2]
    new['Денежные средства'].iloc[counter]= b[3]
    new['Драгоценные металлы'].iloc[counter]= b[4]
    new['Имущество ответственности'].iloc[counter]= b[5]
    new['Крупно-рогатый скот'].iloc[counter]= b[6]
    new['Недвижимость'].iloc[counter]= b[7]
    new['Поручитель'].iloc[counter]= b[8]
    new['Сберегательный депозит'].iloc[counter]= b[9]
    new['Стоимость сельхоз техники'].iloc[counter]= b[10]
    new['Уборка будущего урожая'].iloc[counter]= b[11]
    counter+=1
```
